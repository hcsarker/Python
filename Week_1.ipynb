# 3. Operators
- Arithmetic Operators

  - ``+, -, *, /, //, %, **``

- Comparison Operators

  - ``==, !=, >, <, >=, <=``

- Assignment Operators

  - ``=, +=, -=, *=, /=, %=, **=, //=``

- Logical Operators

  - ``and, or, not``

- Membership Operators

  - ``in, not in``

- Identity Operators

  - ``is, is not``

```python
# Arithmetic Operators
print(3+4)
print(3-4)
print(2*4)
print(4/2)
print(5//2)
print(5%2)
print(2**3) # 2*2*2
```

```python
# Comparison Operators
print(2 == 2) # equal
print(1 != 2)
print(2 > 4)
print(4 < 2)


print(4 >= 5)  # False, False
print(5 <= 5)  # False, True

# !=, >, <, >=, <=
```

```python
# =, +=, -=, *=, /=, %=, **=, //=

a = 2
print(a)
a = a + 1
print(a)

# short
b = 5

b += 6
print(b)
```

```python
# Logical Operators

print(1 == 1 and 'string'=='string')

print(1 == 1 or 'strinG'=='string')

print(not 'string'=='string')

print('----------------------------')
# Membership Operators

print("bangla" in "bangladesh")

print("bangla" not in "Bangladesh")

# in, not in

print('------------------')
# Identity Operators

# is, is not
b = 20

c = 30

b is not c
```

# Coditional statment + For loop + List

```python
for app in apps_data[1:]:
  price = float(app[4])  # type casting/index (list)
  if price == 0.0: # ==, !=, >, <   0.0
    print("This is free App")
  else:
    print("This is premium App")
```

```python
apps_data[18]
```

```python
for index, app in enumerate(apps_data[1:], start=1):
  price = float(app[4])  # type casting/index (list)
  if price == 0.0: # ==, !=, >, <   0.0
    print(app[1], end=' ')   # Facebook 1: This is free App
    print(index, end=': ')
    print("This is free App")
  else:
    print(app[1], end=' ')
    print(index, end=': ')
    print("This is premium App")
```

```python
# number of free and premium application

count_free_apps = 0
count_premium_apps = 0

for app in apps_data[1:]:
  price = float(app[4])  # app[0] -- float("0.0")  -- 0.0
  if price == 0.0:
    count_free_apps = count_free_apps + 1
  else:
    count_premium_apps = count_premium_apps + 1
```

```python
print(count_free_apps)
print(count_premium_apps)
```
