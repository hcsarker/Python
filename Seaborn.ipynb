```python
#pair grid

graph = sns.PairGrid(data = penguins, hue = "species", palette="Set2")
graph.map_upper(sns.scatterplot)
graph.map_lower(sns.kdeplot)
graph.map_diag(sns.histplot)
graph.add_legend()
```
