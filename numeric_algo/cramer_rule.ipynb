```python
import numpy as np
import matplotlib.pyplot as plt
```

```python

# Example system:
# 2x + y - z = 8
# -3x - y + 2z = -11
# -2x + y + 2z = -3
```

```python

A = np.array([[2, 1, -1],
              [-3, -1, 2],
              [-2, 1, 2]])

B = np.array([8, -11, -3])
```

```python

# Determinant of A
D = np.linalg.det(A)
```

```python

if D == 0:
    print("No unique solution (determinant is 0)")
else:
    # Compute Dx, Dy, Dz
    Dx = np.copy(A)
    Dx[:, 0] = B
    Dy = np.copy(A)
    Dy[:, 1] = B
    Dz = np.copy(A)
    Dz[:, 2] = B

    x = np.linalg.det(Dx) / D
    y = np.linalg.det(Dy) / D
    z = np.linalg.det(Dz) / D

print(f"Solution: x = {x:.3f}, y = {y:.3f}, z = {z:.3f}")
```

```python
plt.bar(['x', 'y', 'z'], [x, y, z], color=['blue', 'orange', 'green'])
plt.ylabel('Values')
plt.title('Solution using Cramer\'s Rule')
plt.show()
```
