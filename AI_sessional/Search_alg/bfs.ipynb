```python
import networkx as nx
import matplotlib.pyplot as plt
from collections import deque
```

```python
# Function to perform BFS traversal

def bfs(graph, start):
    visited = set()          # Already visited nodes
    queue = deque([start])   # Queue initialization
    result = []              # Traversal order

    while queue:
        node = queue.popleft()   # Pop from left (FIFO)
        if node not in visited:
            visited.add(node)    # Mark as visited
            result.append(node)  # Store traversal
            # Add neighbors to queue
            for neighbor in graph[node]:
                if neighbor not in visited:
                    queue.append(neighbor)

    return result
```

```python
# Example graph (Adjacency List)
graph = {
    'A': ['B', 'C'],
    'B': ['A', 'D', 'E'],
    'C': ['A', 'F'],
    'D': ['B'],
    'E': ['B', 'F'],
    'F': ['C', 'E']
}
```

```python
# Run BFS
order = bfs(graph, 'A')
print("BFS Traversal Order:", order)
```

```python
# Create Graph using networkx
G = nx.Graph(graph)

# Draw Graph
pos = nx.spring_layout(G)  # Node positioning
nx.draw(G, pos, with_labels=True, node_color="skyblue", node_size=1500, font_size=12, font_weight="bold", edge_color="gray")

# Show traversal order
labels = {node: f"{node}\n({i+1})" for i, node in enumerate(order)}
nx.draw_networkx_labels(G, pos, labels=labels)

plt.title("BFS Graph Visualization")
plt.show()
```
