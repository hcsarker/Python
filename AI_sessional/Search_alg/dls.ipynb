```python
def dls(graph, node, target, limit, depth=0, visited=None):
    if visited is None:
        visited = set()

    # বর্তমান node visit করো
    visited.add(node)
    print(f"Visiting: {node}, Depth: {depth}")

    # যদি target পেয়ে যাই
    if node == target:
        return True

    # যদি depth limit এ পৌঁছে যাই
    if depth >= limit:
        return False

    # Neighbors explore করি (depth + 1)
    for neighbor in graph[node]:
        if neighbor not in visited:
            if dls(graph, neighbor, target, limit, depth + 1, visited):
                return True

    return False

```

```python

# Example Graph
graph = {
    'A': ['B', 'C'],
    'B': ['D', 'E'],
    'C': ['F'],
    'D': [],
    'E': ['F'],
    'F': []
}

# Run DLS
start = 'A'
target = 'F'
limit = 2   # depth limit

found = dls(graph, start, target, limit)

if found:
    print(f"Target {target} found within depth {limit}")
else:
    print(f"Target {target} NOT found within depth {limit}")
```
