```python
import heapq

def ucs(graph, start, goal, costs):
    # Priority Queue (min-heap)
    pq = [(0, [start])]   # (cost, path)
    visited = set()

    while pq:
        cost, path = heapq.heappop(pq)   # সর্বনিম্ন cost এর path নাও
        node = path[-1]

        # Goal check
        if node == goal:
            return path, cost

        if node not in visited:
            visited.add(node)
            for neighbor in graph[node]:
                new_cost = cost + costs.get((node, neighbor), 1)  # edge cost
                new_path = path + [neighbor]
                heapq.heappush(pq, (new_cost, new_path))

    return None, float('inf')
```

```python
graph = {
    'A': ['B', 'C'],
    'B': ['D', 'E'],
    'C': ['F'],
    'D': [],
    'E': ['F'],
    'F': []
}

# Edge Costs (Weighted Graph)
costs = {
    ('A', 'B'): 1, 
    ('A', 'C'): 4,
    ('B', 'D'): 2, 
    ('B', 'E'): 5,
    ('C', 'F'): 1, 
    ('E', 'F'): 1
}

# Run UCS
print("UCS Path:", ucs(graph, 'A', 'F', costs))
```
