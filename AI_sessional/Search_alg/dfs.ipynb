```python
def dfs(graph, node, visited=None, result=None):
    if visited is None:
        visited = set()
    if result is None:
        result = []

    if node not in visited:
        visited.add(node)          # Mark as visited
        result.append(node)        # Store traversal
        for neighbor in graph[node]:
            dfs(graph, neighbor, visited, result)   # Recursive call

    return result

```

```python

# Example graph (Adjacency List)
graph = {
    'A': ['B', 'C'],
    'B': ['A', 'D', 'E'],
    'C': ['A', 'F'],
    'D': ['B'],
    'E': ['B', 'F'],
    'F': ['C', 'E']
}

# Run DFS
print("DFS Traversal:", dfs(graph, 'A'))
```
